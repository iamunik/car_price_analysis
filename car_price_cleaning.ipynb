# WORKING WITH A CAR DATA SET


## CLEANING THE DATA

```python
import pandas as pd
import numpy as np

# I assigned headers because the csv file had no headers within it

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

c_p = pd.read_csv('cars.data', names = headers)
pd.set_option('display.max_columns', None)
c_p
```

```python
# Cleaning missing data from the dataset using the Numpy library we will replace '?' with NaN in the dataset
c_p.replace("?", np.nan, inplace = True)
c_p
```

```python
# Checking for null(NaN) values
# "True" means the value is a missing value while "False" means the value is not a missing value.
missing_data = c_p.isnull()
missing_data
```

```python
# Checking for the locations of missing values and how many they are.
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

Based on the above data
- "normalized-losses": 41 missing data
- "num-of-doors": 2 missing data
- "bore": 4 missing data
- "stroke" : 4 missing data
- "horsepower": 2 missing data
- "peak-rpm": 2 missing data
- "price": 4 missing data

All the intergers below can be replaced with their mean values.
<ul>
    <li>"normalized-losses": 41 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
</ul>
all empty values above should be replaced with its mean.


```python
# Calculate the mean of every non-categorical columns with missing values
avg_norm_loss = c_p["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

avg_bore = round(c_p["bore"].astype("float").mean(axis=0), 2)
print("Average of bore:", avg_bore)

avg_stroke = round(c_p["stroke"].astype("float").mean(axis=0), 2)
print("Average of stroke:", avg_stroke)

avg_horsepower = round(c_p["horsepower"].astype("float").mean(axis=0), 2)
print("Average of horsepower:", avg_horsepower)

avg_peakrpm = round(c_p["peak-rpm"].astype("float").mean(axis=0), 2)
print("Average of peak-rpm:", avg_peakrpm)
```

```python
c_p.describe(include='all')
```

```python
# Replace missing values with the mean
c_p["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
c_p["bore"].replace(np.nan, avg_bore, inplace=True)
c_p["stroke"].replace(np.nan, avg_stroke, inplace=True)
c_p["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
c_p["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)

# Replace door with the most occuring frequency
c_p["num-of-doors"].replace(np.nan, 'four', inplace=True)
c_p
```

```python
# Drop the missing prices row
c_p.dropna(subset=["price"], axis=0, inplace=True)

# Reset the index of the dataset
c_p.reset_index(drop=True, inplace=True)
c_p
```

```python
# Correct the data types of the columns
c_p[["bore", "stroke", "peak-rpm", "price"]] = c_p[["bore", "stroke", "peak-rpm", "price"]].astype("float")
c_p[["normalized-losses"]] = c_p[["normalized-losses"]].astype("int64")

c_p.dtypes
```

>**AFTER CLEANING, WE CAN SEE THAT ;**
>- Columns were added to the dataset for easy identification
>- '?' in the dataset was replaced with NaN
>- NaN rows have been either removed or replaced
>- Data-types have been properly assigned
>- The indexing of the dataset was reset for proper understanding

## NORMALIZING AND BINNING THE DATA

**Normalization** is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling the variable so the variable values range from 0 to 1.<br><br>
Looking at the above dataframe we can scale down **length, width and height**, those are metric values and can be scaled properly.

```python
# I have scaled the values so that it ranges from 0 to 1
c_p['length'] = c_p['length']/c_p['length'].max()
c_p['width'] = c_p['width']/c_p['width'].max()
c_p['height'] = c_p['height']/c_p['height'].max()

# Showing the scaled values
c_p[["length", "width", "height"]].head()
```

**Binning** is the process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.<br>
We can pick different real-values in the datframe to bin since our end product is woking with the prices, but we know that picking prices of car also depends mainly on the cars **horsepower** so we will bin the **horsepower** and reduce it fro **59** unique variable to **3** unique variables, that is; **low, medium and high horsepower**<br><br>
Before binning the column we have to change its data-type to integer since it's currently an object.

```python
# Converted this to an interger from being an object
c_p['horsepower'] = c_p['horsepower'].astype(int, copy=True)

# Using the numpy package we can group this into 3 "low", "medium", "high"
bins = np.linspace(min(c_p["horsepower"]), max(c_p["horsepower"]), 4)

# Set its group name
group_names = ['Low', 'Medium', 'High']

# Create a new column holding the values of the binned data
c_p['horsepower-binned'] = pd.cut(c_p['horsepower'], bins, labels=group_names, include_lowest=True )
c_p[['horsepower','horsepower-binned']].head(20)
```

```python
c_p["horsepower-binned"].value_counts()
```

We can now plot an Histogram distribution for each of the binned values

```python
%matplotlib inline
import matplotlib.pyplot as plt


plt.bar(group_names, c_p["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

>At the end of the section:<br>
>- We scaled important datasets
>- We were able to also group some data thus reducing it from 59 unique variables to just 3.
>- Before grouping we converted a column to its correct data types
>- We plotted a bar chart to see what we did

## CREATING DUMMY VARIABLES

We are creating dummy variables because we need to use categorical data in our regression analysis, it is called dummy because the values in it are just 1's and 0's and the numbers themselves have no real meaning

```python
c_p.describe(include='all')
```

> The **fuel-type** and **aspiration** columns has 2 unique values and will be used during the regression analysis

```python
dummy_fuel = pd.get_dummies(c_p['fuel-type'])
dummy_fuel.head()
```

```python
dummy_fuel.rename(columns={"gas": "fuel-gas", "diesel":"fuel-diesel"}, inplace=True)
dummy_fuel.head()
```

```python

dummy_aspirate = pd.get_dummies(c_p['aspiration'])

dummy_aspirate.rename(columns={"std":"aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)

dummy_aspirate.head()
```

After creating a dummy column we can now join it to our existing dataframe

```python
# Merging 'c_p' and 'dummy_fuel'
c_p = pd.concat([c_p, dummy_fuel], axis=1)

# Merging 'c_p' and 'dummy_aspirate'
c_p = pd.concat([c_p, dummy_aspirate], axis=1)

# Drop the column 'fuel-type'
c_p.drop('fuel-type', axis=1, inplace=True)

# Drop the original column 'aspirate'
c_p.drop('aspiration', axis=1, inplace=True)
```
